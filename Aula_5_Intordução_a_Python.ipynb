```python
# criação uma calculadora de desconto em Python para ajudar os vendedores a calcularem o valor final de uma compra com base no preço do produto e em um desconto percentual oferecido.

valor_produto = float(input("valor do produto:"))
```

```python
porcentagem_desconto = float(input("porcentagem do desconto:"))
```

```python
#Convertento variaveis em float
valor_produto = float(valor_produto)
porcentagem_desconto = float(porcentagem_desconto)
```

```python
#Verifica se o desconto esta dentro dos limites aceitaveis
if porcentagem_desconto <= 0 or porcentagem_desconto >= 100:
    print("porcentagem de desconto invalida")
else: #Calcular valor do desconto
  desconto = valor_produto * (porcentagem_desconto / 100)
  print(desconto)
```
