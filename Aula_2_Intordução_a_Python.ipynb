```python
#Algritimo de recomendação para cinema


idade = int(input("Digite sua idade:"))
```

```python
#Condições para escola do filme de acordo com a idade
# < de 12 e > 0 Filme 1
# < de 18 e > 12 filme 2
# <= 18 filme 3

if idade  <= 12 and idade > 0 :
  print("recomenda-se o fiime 1")
elif idade < 18 and idade > 12 :
  print("recomenda-se o filme 2")
elif idade >= 18 and idade <= 100:
  print("Recomenda-se o filme 3")
else:
  print("idade invalida")
```
