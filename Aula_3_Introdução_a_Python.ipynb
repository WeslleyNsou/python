```python
#Algoritimo para classificação de filmes
#O programa precisa percorrer todos os filmes e atribuir uma nota a cada um deles
#E importante adicionar uma maneira de para o programa

filmes = [1, 2, 3]
```

```python
for filme in filmes:
    #Solicita uma classificação
      classificacao = int(input(f"Classifique o filme {filme} de 1 a 5 (ou digite 0 para parar):"))
      #Verifica se o usário deseja parar o programa
      if classificacao == 0:
          print("Classificação finalizada")
          break
      #Verifica se a classificaço esta dentro do intervalo valido
      elif classificacao < 1 or classificacao >5:
          print("Classificação inválida")
          break
      else:
          print("Classificação realizada com sucesso")


```
