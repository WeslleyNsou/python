```python
import re # Importar a biblioteca re para usar expressões regulares
```

```python
#Classe Pacientes
class Paciente:
    def __init__(self, nome, idade, numero):
        self.nome = nome
        #Validaçao da idade
        self.idade = idade
        # Adicionar validação e formatação para o número aqui
        self.numero = self.formatar_numero(numero)

    def formatar_numero(self, numero):
        # Remover todos os caracteres não numéricos
        numero_limpo = re.sub(r'\D', '', numero)

        # Aplicar a formatação (DDD) DDDD-DDDD
        if len(numero_limpo) == 10:
            # Formato para números com 10 dígitos (sem o 9)
            return f"({numero_limpo[:3]}) {numero_limpo[3:6]}-{numero_limpo[6:]}"
        elif len(numero_limpo) == 11:
            # Formato para números com 11 dígitos (com o 9)
            return f"({numero_limpo[:3]}) {numero_limpo[3:7]}-{numero_limpo[7:]}"
        else:
            # Manter o número original se não tiver 10 ou 11 dígitos
            print(f"Aviso: Número '{numero}' não pôde ser formatado corretamente.")
            return numero_limpo # Retornar o número limpo sem formatação

    def __str__(self):
        return f"Nome: {self.nome}\nIdade: {self.idade}\nNumero: {self.numero}"
```

```python
#Lista para armazenar os pacientes
lista_pacientes = []
```

```python
#Cadastro de Pacientes
def add_paciente(nome, idade, numero):
    paciente = Paciente(nome, idade, numero)
    lista_pacientes.append(paciente)
    print(f"Paciente '{nome}' cadastrado com sucesso!")
```

```python
def listar_pacientes():
    if len(lista_pacientes) == 0:
        print("Não há pacientes cadastrados.")
    else:
        print("Lista de Pacientes:")
        for paciente in lista_pacientes:
            print(paciente)
```

```python
len(lista_pacientes)#Numero de pacientres cadastrados
```

```python
#Idade media dos pacientes
def calcular_idade_media():
  idade_total = 0
  if len(lista_pacientes) > 0:
      for paciente in lista_pacientes:
          idade_total += paciente.idade
      idade_media = idade_total / len(lista_pacientes)
      print(f"\nA idade média dos pacientes é: {idade_media:.2f} anos")
  else:
      print("Não há pacientes cadastrados para calcular a idade média.")
```

```python
#Paciente mais novo e mais velho
if len(lista_pacientes) > 0:
    paciente_mais_novo = lista_pacientes[0]
    for paciente in lista_pacientes:
        if paciente.idade < paciente_mais_novo.idade:
            paciente_mais_novo = paciente

    paciente_mais_velho = lista_pacientes[0]
    for paciente in lista_pacientes:
        if paciente.idade > paciente_mais_velho.idade:
            paciente_mais_velho = paciente

    print(f"O paciente mais novo é:\n{paciente_mais_novo.nome}")
    print(f"O paciente mais velho é:\n{paciente_mais_velho.nome}")
else:
    print("Não há pacientes cadastrados para determinar o mais novo ou mais velho.")
```

```python
#Pesquisa o paciente pelo nome
def pesquisar_paciente():
  nome_pesquisa = input("Digite o nome do paciente que deseja pesquisar: ")
  paciente_encontrado = None
  for paciente in lista_pacientes:
      if paciente.nome == nome_pesquisa:
          paciente_encontrado = paciente
          break # Adicionado para parar a busca assim que o paciente é encontrado
  if paciente_encontrado:
    print(paciente_encontrado)
  else:
    print(f"Paciente '{nome_pesquisa}' não encontrado.")
```

```python
#Menu
while True:
  print("=== SISTEMA CLÍNICA VIDA+ ===")
  print("1. Cadastrar paciente")
  print("2. Ver estatísticas")
  print("3. Buscar paciente")
  print("4. Listar todos os pacientes")
  print("5. Sair")

  Escolha = (input("Escolha uma opção: "))

  if Escolha == "1":
    nome = input("Digite o nome do paciente: ")
    idade = int(input("Digite a idade do paciente: "))
    numero = input("Digite o número do paciente: ")
    add_paciente(nome, idade, numero)
    print("Paciente cadastrado com sucesso!")

  elif Escolha == "2":
      if len(lista_pacientes) > 0:
          calcular_idade_media()
          print(f"\nO paciente mais velho é: {paciente_mais_velho.nome}")
          print(f"\nO paciente mais novo é: {paciente_mais_novo.nome}\n")
      else:
          print("\nNão há pacientes cadastrados para determinar o mais novo ou mais velho.\n")


  elif Escolha == "3":
      pesquisar_paciente()

  elif Escolha == "4":
      listar_pacientes()

  elif Escolha == "5":
      print("Saindo do sistema...")
      break

  else:
      print("Opção inválida. Tente novamente.")
```
