```
#Automatizando média de notas com Python

nota_1 = int(input("Digite sua primeira nota:"))
nota_2 = int(input("Digite sua segunda nota:"))
nota_3 = int(input("Digite sua terceira nota:"))
nota_4 = int(input("Digite sua ultima nota:"))
```

```
#Somando notas para definir media

media = (nota_1 + nota_2 + nota_3 + nota_4)
```

```
#Estutura condicional para definir se o aluno foi aprovado ou reprovado

if media >= 20 :
  situacao = "Aprovado!"
else :
  situacao = "Reprovado!"
```

```
#Exibição

print(media)
print(situacao)
```
