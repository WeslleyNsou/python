```python
import matplotlib.pyplot as plt
#Classe para representar um livro
class Livro:
    def __init__(self, titulo, autor, genero, quantidade_disponivel):
      self.titulo = titulo
      self.autor = autor
      self.genero = genero
      self.quantidade_disponivel = quantidade_disponivel

    def __str__(self):
        return f"{self.titulo} - {self.autor} - {self.genero} - {self.quantidade_disponivel}"
```

```python
#lista para armazenar os livros
Biblioteca = []

#Lista para armazenar os genros dos livros
generos = []

#Lista para armazenar a quantidade de livros
quantidade_livros = []

#Função para adicionar um novo livro a lista Biblioteca
def adicionar_livro(titulo, autor, genero, quantidade_disponivel):
  novo_livro = Livro(titulo, autor, genero, quantidade_disponivel)
  Biblioteca.append(novo_livro)
  generos.append(genero)
  quantidade_livros.append(quantidade_disponivel)
  print(f"Livro '{titulo}' adicionado com sucesso! Ele possui {quantidade_disponivel} unidades disponiveis")

#Função para listar os Livros na Biblioteca
def listar_livros():
  print("Livros disponiveis na Biblioteca")
  for livro in Biblioteca:
    print(livro)

#Adicionando alguns livros na bilbioteca
adicionar_livro("Dom Quixote", "Miguel de Cervantes", "Aventura", 160)
adicionar_livro("Orgulho e Preconceito", "Jane Austen", "romance", 181)
adicionar_livro("1984", "George Orwell", "Ação", 194)
adicionar_livro("A Revolução dos Bichos", "George Orwell", "ficção", 194)
```

```python
print(quantidade_livros)
print(generos)
```

```python

#Cria uma Grafico de livros por genero

plt.bar(generos, quantidade_livros)
plt.xlabel('Genero')
plt.ylabel('Quantidade de Livros')
plt.title('Quantidade de Livros por Genero')
plt.show()
```
